```python
import numpy as np
import pandas as pd
import csv
```

For 1 conc. contains 300 data-set of x,y (x1, y1, x2, y2, ..., x300, y300)

```python
df = pd.read_csv('20m_poly.csv')
df
```

Extracting x and y values for each dataset

```python
num_datasets = 300
x_values = []
y_values = []
for i in range(1, num_datasets + 1):
    x_col = f'x{i}'
    y_col = f'y{i}'
    x_values.append(df[x_col])
    y_values.append(df[y_col])
```

Fitting polynomials for each dataset

```python
polynomials = []
for x, y in zip(x_values, y_values):
    degree = len(x) - 1         ##maximize 
    coefficients = np.polyfit(x, y, degree)
    polynomial = np.poly1d(coefficients)
    polynomials.append(polynomial)
```

Import and arrange format to csv

```python
# Write polynomial equations along with their coefficients to a CSV file
with open('20m_poly_new.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    # Writing header row
    header_row = ['Dataset', 'Polynomial Equation'] + [f'Coefficient {i}' for i in range(len(polynomials[0].coefficients))]
    writer.writerow(header_row)
    
    # Writing polynomial equations and coefficients for each dataset
    for i, polynomial in enumerate(polynomials, start=1):
        coefficients = list(polynomial.coefficients)
        row = [f'Dataset {i}', str(polynomial)] + coefficients
        writer.writerow(row)

print("Polynomial equations along with their coefficients have been saved to '20m_poly_new.csv'.")
```
